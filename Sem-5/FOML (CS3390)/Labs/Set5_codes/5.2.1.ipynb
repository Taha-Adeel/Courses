```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import normalize
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
```

# Read

```python
xls = pd.ExcelFile('LSVT_voice_rehabilitation.xlsx') 
df1 = pd.read_excel(xls, 'Data')
df2 = pd.read_excel(xls, 'Binary response')

print(df1.shape, df2.shape)

# checking for missing data
c = df1.count()
for i in c: # prints how many non-NAN values are there for each column
    assert i==df1.shape[0]
```

# Converting to numpy

```python
X = df1.to_numpy()
y = (df2.to_numpy()).ravel()
```

```python
X = normalize(X)
```

```python
for reg in [0.01, 0.1, 1, 10, 100]:
    print("For Regularization ", reg)
    clf_logr = LogisticRegression(C=1.0/reg)
    net_logr = []
    for i in range(5):
        kf = KFold(n_splits=3, random_state=i, shuffle=True)
        
        for train_index, val_index in kf.split(X):
            X_train, X_val = X[train_index], X[val_index]
            Y_train, Y_val = y[train_index], y[val_index]
            
            clf_logr.fit(X_train, Y_train)
            y_pred = clf_logr.predict(X_val)
            
            net_logr.append(clf_logr.score(X_val, Y_val))
    print(" Avg accuracy score", np.mean(net_logr))
```
