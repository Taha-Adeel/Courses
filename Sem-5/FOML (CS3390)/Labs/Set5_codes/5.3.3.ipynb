# Code for **training** SVM classifier (for a give hyperparameter configuration) and **plotting** the decision boundary.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets

iris = datasets.load_iris()
X = iris.data[:, :2]  # taking only the 1st 2 features so that we can show decision boundary with these 
y = iris.target

C = 1.0  # SVM regularization parameter
clf = svm.SVC(kernel='linear', C=C).fit(X, y)

# In our data, we note what is the range of sepal length & sepal width features. 
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
# Then we get a 2D mesh with (x, y) points where x points are sepal lengths falling in the range found earlier, y points are sepal widths falling in the range found earlier.  
h = 0.02
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))


# reshaping & concatenating the points as per the classifier's requirement.
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape) # reshaping to plot
# as the input is 2D, the graph of input-output will be in 3D, hence we are plotting its contours.
plt.contourf(xx, yy, Z) # this plots contour of the classifier 
plt.scatter(X[:, 0], X[:, 1], c=y) # to visualize, how pts got classified
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max()) #x-axis limits to be shown
plt.ylim(yy.min(), yy.max()) #y-axis limits to be shown
plt.xticks(())
plt.yticks(())
plt.title('SVC with linear kernel')

plt.show()
```

You can similarly run it for different hyperparameters.
